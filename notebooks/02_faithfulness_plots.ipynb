# Faithfulness Evaluation Plots

This notebook visualizes and analyzes the faithfulness results of the GNN explanation methods.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import numpy as np

results_path = Path("../results/faithfulness_results.csv")
if results_path.exists():
    df = pd.read_csv(results_path)
    print("=== Loaded Real Faithfulness Results ===")
    print(df)
else:
    # Populate simulated results for visualization purposes
    data = {
        'method': ['attention', 'gnnexplainer', 'pgexplainer', 'kec'],
        'sufficiency': [0.65, 0.78, 0.82, 0.89],
        'necessity': [0.12, 0.28, 0.35, 0.48],
        'fidelity_plus': [0.15, 0.30, 0.38, 0.52],
        'fidelity_minus': [0.35, 0.22, 0.18, 0.11],
        'sparsity': [0.45, 0.25, 0.20, 0.15],
        'num_evaluated': [200, 200, 200, 200]
    }
    df = pd.DataFrame(data)
    print("=== Loaded Demonstration Faithfulness Results ===")
    print(df)
```

## 1. Sufficiency & Necessity Comparison

Sufficiency measures if the explanation alone preserves prediction accuracy (higher is better).
Necessity measures if removing the explanation decreases prediction accuracy (higher is better).

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Set custom theme style
sns.set_theme(style="whitegrid")

# Sufficiency
sns.barplot(data=df, x='method', y='sufficiency', ax=axes[0], palette='viridis', edgecolor='black')
axes[0].set_title('Explanation Sufficiency (Higher is Better)', fontsize=14)
axes[0].set_ylabel('Sufficiency Score', fontsize=12)
axes[0].set_xlabel('Explanation Method', fontsize=12)
axes[0].set_ylim(0, 1.0)
for p in axes[0].patches:
    axes[0].annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height() - 0.08),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', color='white', weight='bold')

# Necessity
sns.barplot(data=df, x='method', y='necessity', ax=axes[1], palette='magma', edgecolor='black')
axes[1].set_title('Explanation Necessity (Higher is Better)', fontsize=14)
axes[1].set_ylabel('Necessity Score', fontsize=12)
axes[1].set_xlabel('Explanation Method', fontsize=12)
axes[1].set_ylim(0, 1.0)
for p in axes[1].patches:
    axes[1].annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height() - 0.08),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', color='white', weight='bold')

plt.tight_layout()
plt.show()
```
